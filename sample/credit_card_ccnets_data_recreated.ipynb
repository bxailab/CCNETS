# DataLoader

df.isnull().sum()

```python
# https://fraud-detection-handbook.github.io/fraud-detection-handbook/Chapter_7_DeepLearning/FeedForwardNeuralNetworks.html
class Dataset(torch.utils.data.Dataset):
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __len__(self):
        return len(self.x)

    def __getitem__(self, index):
        vals = torch.tensor(self.x[index], dtype = torch.float32)
        label = torch.tensor(self.y[index], dtype= torch.float32).unsqueeze(-1)
        return vals, label
```

```python
y = df[['Class']]
X = df.drop(['Class'],axis=1)

sc = RobustScaler()
X['scaled_amount'] = sc.fit_transform(X['Amount'].values.reshape(-1,1))
X['scaled_time'] = sc.fit_transform(X['Time'].values.reshape(-1,1))
X.drop(['Time','Amount'], axis=1, inplace=True)
X = X[:]
```

# Training

```python
from ccnets.config import get_parser
from ccnets.ccnets import CCNets
from nn.custom_deepfm import DeepFM
from nn.custom_dnn import ResMLP, MLP 
from ccnets.utils.log import create_log_details, create_log_name
from ccnets.utils.setting import set_random_seed

import os
from torch.utils.tensorboard import SummaryWriter
```

```python
args = get_parser()
args.device = torch.device('cuda:0' if (torch.cuda.is_available() and args.ngpu > 0) else "cpu")
```

```python
file_path =  os.path.abspath('')
from pathlib import Path
file_name = Path(file_path).stem
model_path = path_append + f"models/{file_name}/"
temp_path = path_append + f"models/{'temp_'}{file_name}/"
log_path = path_append + f"log/{file_name}/"

if Path(temp_path).exists() is False: 
    os.mkdir(temp_path)

if Path(model_path).exists() is False: 
    os.mkdir(model_path)

if Path(log_path).exists() is False: 
    os.mkdir(log_path)  

args.model_path = model_path
args.temp_path = temp_path
```

```python
args.num_epoch = 32
args.lr = 2e-4
args.batch_size = 64
args.step_size = 10

args.num_layer = 3
args.hidden_size = 256

args.obs_size = 30
args.label_size = 1
args.explain_size = 26   #  10% compression(obs 30, label 1, explain 26)
args.seq_len = 0

args.num_checkpoints = 1335
args.use_one_hot = False

args.reasoner_joint_type = "add"
args.producer_joint_type = "add"
args.label_type = "UC" 

args.loss_reduction = "all"
args.error_reduction = "none"

args.loss_type = "L1"
args.error_type = "L1"

args.explain_fn = "sigmoid"  # Explainer final layer
args.label_fn = "none"       # Reasoner final layer
args.obs_fn = "none"         # Producer final layer
```

```python
# 30 train : 70 test without shuffle to have a clear difference since the dataset is quite easy to master
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.7, shuffle= False) 

X_train = X_train.iloc[:, :].values 
X_test = X_test.iloc[:, :].values 
y_train = y_train.iloc[:, -1].values
y_test = y_test.iloc[:, -1].values

trainset = Dataset(X_train, y_train)
testset = Dataset(X_test, y_test)
```

```python
from ccnets.supervised_learning_with_ccnets import SupervisedLearningWithCCNets
```

```python
args.model_path = model_path

log_details = create_log_details(args)
args.log = SummaryWriter(log_dir=create_log_name(log_path, log_details))
set_random_seed(0)

ccnets = CCNets(args, MLP, DeepFM, ResMLP)
#ccnets.load_models()
ccnets.train(trainset, testset)  
```

```python
from sklearn.metrics import classification_report 
import warnings
from sklearn.exceptions import ConvergenceWarning
from sklearn.metrics import log_loss
warnings.filterwarnings('ignore', category=ConvergenceWarning)
```

```python
def get_classification_report(inferred_y, target_y, label_size):
    target_y = target_y
    inferred_y = inferred_y
    option = "1" if label_size <= 2 else "macro avg"

    if label_size < 2:
        target_y = target_y
        inferred_y = [1 if prob > 0.5 else 0 for prob in inferred_y]      
        report = classification_report(target_y, inferred_y, output_dict=True, labels = list([0, 1]), zero_division=True)[option]
    else:
        target_y = target_y.argmax(dim=-1)
        inferred_y = inferred_y.argmax(dim=-1)
        report = classification_report(target_y, inferred_y, output_dict=True, labels = list(range(label_size)), zero_division=True)[option]
    precision = report["precision"]
    recall = report["recall"]
    f1 = report["f1-score"]
    
    return np.array([precision, recall, f1])

def print_metric(label_type, matric):

    if label_type == "UC":
        precision, recall, f1 = matric
        print('precision: %.4f\trecall: %.4f\tf1: %.4f'
            % (precision, recall, f1))
        
def log_metric(log, label_type, iters, matric):
    if label_type == "UC":
        precision, recall, f1 = matric
        log.add_scalar("Test/precision", precision, iters)
        log.add_scalar("Test/recall", recall, iters)
        log.add_scalar("Test/f1", f1, iters)
        
def log_train(log, iters, inference_loss):
    log.add_scalar("Train/inference_loss", inference_loss, iters)
    
def log_test(log, iters, inference_loss):
    log.add_scalar("Test/inference_loss", inference_loss, iters)
```

```python
from tqdm.notebook import tqdm_notebook
from sklearn.neural_network import MLPClassifier
import numpy as np
####################################################################
################### reference at cooperative_net.py ################
####################################################################
# supevised learing modelding from recreated datasets
# sl: supervised_learning using original dataset
# generate: supervised_learning using ccnets generate function
# generate_with_random_label: supervised_learning using ccnets generate_with_random_label function
# generate_with_random_explain: supervised_learning using ccnets generate_with_random_explain function
# reconstruct: supervised_learning using ccnets reconstruct function
# reverse_label: supervised_learning using ccnets reverse_label function
# balance_label: supervised_learning using ccnets balance_label function
####################################################################
####################################################################
####################################################################


args.supervised_learning_loss_type = "MSE"
```

```python
data_recreation_types = ["s1"]

for data_type in data_recreation_types:
    args.model_path = path_append + f"models/" + data_type + "/"  
    if Path(args.model_path).exists() is False: 
        os.mkdir(args.model_path)
        
    log_details = create_log_details(args)
    log = SummaryWriter(log_dir=create_log_name(log_path, data_type))
    set_random_seed(0)
    
    sl_with_ccnets = SupervisedLearningWithCCNets(args, MLP)
    x,y = sl_with_ccnets.train(trainset, testset, ccnets, data_type)

    model = MLPClassifier((256,3),max_iter=1, batch_size = 64,warm_start = True,random_state = 100)
    for epoch in tqdm_notebook(range(args.num_epoch)):
        #train model
        model.fit(x,y.reshape(-1))

        #test model to test data
        pred_y = model.predict(X_test)
        metrics = get_classification_report(pred_y, y_test, 1)
        print_metric("UC", metrics)
        log_metric(log, "UC", epoch, metrics)
        log_train(log, epoch, model.loss_)
        log_test(log,epoch,log_loss(y_test,pred_y.astype(int)))
        log.flush()


    
```
